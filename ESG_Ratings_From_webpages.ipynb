```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Create an empty data frame with the desired column names
df = pd.DataFrame(columns=['Company Name', 'Sustainalytics', 'Refinitive', 'S&P Global'])

# Define the list of company names
company_names = ['Schaeffler AG', 'Denso Corp', 'Volkswagen AG', 'Continental AG']

# Set up Selenium driver (assuming you have already installed and set up Selenium and the appropriate web driver)
driver = webdriver.Chrome()  # Change this to the path of your web driver if needed

# Loop through the company names
for i, company_name in enumerate(company_names):
    ######### First Website: Sustainalytics
    driver.get('https://www.sustainalytics.com/esg-rating')
    time.sleep(10)
    
    # Dismiss cookie confirmation on the page  if it appears
    try:
        driver.find_element(By.ID, 'hs-en-cookie-confirmation-buttons-area').click()
    except:
        pass
    
    # Fill in the company name in the form form
    form_input = driver.find_element(By.XPATH, '/html/body/main/div/div/div/section[1]/div/div/div[1]/input')
    form_input.clear()
    form_input.send_keys(company_name)
    time.sleep(10) 
    
    # Click on the element
    driver.find_element(By.XPATH, '//*[@id="searchResults"]/div/div/div/a/div/div[1]/div').click()

    # Wait for the score to load
    time.sleep(10)  # Adjust the wait time as needed
    
    # Retrieve the score
    score1 = driver.find_element(By.XPATH, '//*[@id="Contentplaceholder1_C001_Col00"]/div[1]/section[1]/div/div[1]/div[1]/div[5]/div[1]/div[2]/div[1]/span').text
    
    ######### Second Website: Refinitive
    driver.get('https://www.refinitiv.com/en/sustainable-finance/esg-scores')
    time.sleep(10)
    
    # Dismiss cookie confirmation if it appears
    try:
        driver.find_element(By.XPATH, '//*[@id="onetrust-reject-all-handler"]').click()
    except:
        pass
     
    # Fill in the form
    form_input = driver.find_element(By.XPATH, '//*[@id="searchInput-1"]')
    form_input.clear()
    form_input.send_keys(company_name)
    time.sleep(10) 
    
    # Click on the element
    driver.find_element(By.XPATH, '//*[@id="esg-data-body"]/div[1]/div/div/div[1]/div/button[2]').click()

    # Wait for the score to load
    time.sleep(10)  # Adjust the wait time as needed
    
    # Retrieve the score
    score2 = driver.find_element(By.XPATH, '//*[@id="esg-data-body"]/div[2]/div/div/div/div/div/div[1]/div/div/div[1]/h3/strong').text
    
    ######### Third Website: S&P Global
    driver.get('https://www.spglobal.com/esg/scores/results')
    time.sleep(10)
    
    # Dismiss cookie confirmation if it appears
    try:
        driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]').click()
    except:
        pass
     
    # Fill in the form
    form_input = driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/div[3]/div/div/div[2]/div[1]/input')
    form_input.clear()
    form_input.send_keys(company_name)
    time.sleep(10) 
    
    # Click on the element
    driver.find_element(By.XPATH, '/html/body/div[3]/div[1]/div[3]/div/div/div[2]/div[2]/ul/li/a').click()

    # Wait for the score to load
    time.sleep(10)  # Adjust the wait time as needed
    
    # Retrieve the score
    score3 = driver.find_element(By.XPATH, '//*[@id="esg-score"]').text
    
    df.loc[i] = [company_name, score1, score2, score3]  # Update the appropriate column

# Close the driver
driver.quit()

# Display the final data frame
df.head()
```
